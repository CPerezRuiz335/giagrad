```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

%matplotlib inline
import numpy as np
import sys
 
# setting path
sys.path.append('../.')
from giagrad.tensor import Tensor
```

```python
from giagrad.display import draw_dot
```

```python
a = Tensor([[-1.,-2.],[3.,-4.],[-5.,6.]], 
          requires_grad=True, name='a')

b = a.mean()
b.name = 'b'

a2 = a.exp()
a2.name = "a'"

c = Tensor([[2.,2.],[2.,2.],[3.,3.]], 
          requires_grad=True, name='c')

c = c * a2.log().relu()
c.name = "c'"

d = b * c
d.name = 'd'

e = Tensor([[5.,5.,5.]],
          requires_grad=True, name='e')

f = e @ d
f.name = 'f'

g = f.sum()
g.name = 'g'
```

```python
draw_dot(g, rankdir='TB')
```

```python
g.backward(debug=True)
draw_dot(g, rankdir='TB')
```
