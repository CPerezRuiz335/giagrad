```python
from giagrad.display import draw_dot
```

```python
a = np.array([[2.,2.],
              [2.,2.],
              [3.,3.]])

b = np.array([[1.,2.],
              [3.,4.],
              [5.,6.]])
a_tensor = Tensor(a, requires_grad=True)
a_tensor.name = 'a'
c_tensor = Tensor(b, requires_grad=True)
c_tensor.name = 'c'
```

```python
b_tensor = a_tensor.min()
b_tensor.name = 'b'
d_tensor = b_tensor * c_tensor
d_tensor.name = 'd'
d_tensor
```

```python
draw_dot(d_tensor)
```
